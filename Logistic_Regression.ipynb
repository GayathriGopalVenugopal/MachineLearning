# **LOGISTIC REGRESSION**

Logistic regression is a statistical analysis method used to predict a data value based on prior observations of a data set.
In regression analysis, logistic regression is estimating the parameters of a logistic model in order to find the probability of event=True and event=False.

This regression method is used when the dependent variable is binary (0/ 1, True/ False, Yes/ No) where the value of Y ranges from 0 to 1 and it can represented by following equation where p is the probability of presence of the characteristic of interest


> **odds= p/ (1-p) = probability of event occurrence / probability of no event occurrence**

> **ln(odds) = ln(p/(1-p))**

> **logit(p) = ln(p/(1-p)) = b0 + b1X1 + b2X2 + b3X3....+ bkXk**

# **Dataset-1 : Synthetic Dataset**

Synthetic data is generally created artificially rather than generated from real events. This type of data is mainly used in AI model, Machine learning and for developing and testing new products. One of the main differences of synthetic data and actual data is that synthetic data can be used to meet some specific needs or conditions which is not possible in real data.

Some of the features of synthetic data includes:

· Real data has some data usage restrictions due to some rules and regulations but synthetic data can replicate all properties without exposing the real data.

· In some cases, real data may not exist so synthetic data has to be created to study the scenario.

· Synthetic data mainly focuses on multivariate relationship.



This synthetic dataset contains four classes and 80 samples per classes. The data is first plotted and split in 3:4 ratio for taking training and testing datasets. Finally, decision boundary plots are also drawn.
   
##  **Method:**


1.The synthetic dataset is imported from the sklearn.datasets library.

2.The datatype is then converted to pandas dataframe for easy usage of the data.

3.Then the data is classified into 4 classes and 80 samples per classes

4.Each class is classified into 4 different colors, for example, red, blue, green and yellow.

5.Scatter plots are used for each class to check the relation of class and each variable values.

6.The dataset is then split into training and testing dataset. 75% of the dataset is taken as training set and the remaining 25% is taken as the testing set.

7.After this, the model is then trained using Logistic Regression available in Scikitlearn library

The accuracy of the model is then predicted using the test data

```python
# import libraries
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from pandas import DataFrame
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt
import numpy as np
```

```python
## randomly generate a 2d, 4 classes dataset
# run several times the cell to observe the results 
X1, y1 = make_classification(n_samples=320, n_classes=4, n_features=2, n_redundant=0, 
                            n_informative=2,
                            n_clusters_per_class=1)
#create a dataframe
df = DataFrame(dict (x=X1[:,0], y=X1[:,1], label=y1))
# four classes 
colors = {0: 'red' , 1:'blue' , 2:'green', 3:'yellow' }
# figure
fig, ax = plt.subplots()
grouped = df.groupby('label')
# scatter plot
for key, group in grouped:
  group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key]) 
# show the plot
plt.show()
```

```python
# show first 15 rows 
df.head(15)
```

```python
# count the data elements for each class
df['label'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
# training: 75% (0.75), test: 25% (0.25)
X_train,X_test,y_train,y_test = train_test_split(X1, y1, test_size=0.25)

 
# Logistic Regression model
LR_model = LogisticRegression()
# fit to data a Logistic regression model
LR_model.fit(X_train, y_train)
```

```python
# Plotting the decision boundary 
# from the LogisticRegression model
plot_decision_regions(X_train, y_train, clf=LR_model, legend=2)
```

```python
# make predictions (assign class labels)
y_pred = LR_model.predict(X_test)
# show the inputs and predicted outputs
for i in range(len(X_test)):
    #print("X=%s, Class Predicted = %s" % (X_test[i], y_pred[i]))
  print("X{0} = {1}, Class Predicted = {2}".format(i, X_test[i], y_pred[i]))
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
con=confusion_matrix(y_test,y_pred)
print(con)
print(accuracy_score(y_test,y_pred))
```

```python
# Plotting the new decision boundary 
# from the LogisticRegression model
plot_decision_regions(X_test, y_pred, clf=LR_model, legend=2)
```

Conclusion:
The Logistic Regression model has been implemented on Synthetic data set. The model provides an accuracy of around 91% in prediction using test data.

# **Dataset-3 : Wine Dataset**

The Wine dataset tabulates the results and observations of the chemical analysis performed on wine distilled around a region in Italy, from three different cultivators.
The thirteen constituents that were mainly focused on, are as follow:
1. Alcohol
2. Malic acid
3. Ash
4. Alkalinity of ash
5. Magnesium
6. Total phenols
7. Flavanoids
8. Nonflavanoid phenols
9. Proanthocyanins
10.Color intensity
11.Hue
12.OD280/OD315 of diluted wines
13.Proline

There are no missing values for any of these variables in the dataset.

**Method**:

1. The wine dataset was first imported from the sklearn.datasets library.
2. The dataset was retrieved in Bunch datatype format
3. The Bunch datatype was then converted to pandas dataframe for easy usage of the data.
4. Barplots were plotted for each variable with class variable on the independent axis using seaborn library to understand the variation in each variables according to class of wine
5. Histograms were also plotted for each of these variables to check the relation of class and each variable values to compare with those obtained from the barplot.
6. The dataset was then split into training and testing dataset. 80% of the dataset was taken as training set and the remaining as testing set.
7. The training dataset was then standardised using StandardScaler to bring multiscaled variables to a common scale for providing better accuracy.
8. After standardisation, the model is then trained using Logistic Regression available in Scikitlearn library
9. The accuracy of the model is then predicted using the test data

```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression
from mlxtend.plotting import plot_decision_regions
```

```python
# import wine dataset from sklearn
from sklearn.datasets import load_wine
data = load_wine()
#checking the datatype of the dataset
type(data)
```

```python
#create a dataframe
df = pd.DataFrame(data.data, columns=data.feature_names)
df['class'] = pd.Series(data.target)
# show first 5 rows 
df.head()
```

```python
df['class'].value_counts()
```

```python
# provides information about attributes
df.info()
```

```python
# drop column 'class' from dataframe
wine_data=df.drop(columns='class', errors='ignore')
```

```python
# plot data using Bargraph
for wd in wine_data:
  fig=plt.figure(figsize=(10,8))
  sns.barplot(x=df['class'], y=df[wd])
```

```python
# plot data using histogram
for wd in df:
  df.hist(column=wd,color='#86bf91',figsize=(10,8))
```

```python
X=df.drop(columns='class')
Y=df['class']
# selecting particular rows and columns from a set of values of a dataframe
X2 = X.iloc[:,[0, 12]]
# training: 80% (0.80), test: 20% (0.20)
X_train,X_test,Y_train,Y_test=train_test_split(X2,Y,test_size=0.20, random_state=0)
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

```python
#Standardise data for better results
scalar=sklearn.preprocessing.StandardScaler()
X_train=scalar.fit_transform(X_train)
X_test=scalar.fit_transform(X_test)
```

```python
# Logistic Regression model
regressor=LogisticRegression()
# fit to data a Logistic regression model
regressor.fit(X_train,Y_train)
```

```python
# changing the variable datatype from series to array 
y_train=np.array(Y_train)
y_test=np.array(Y_test)
# checking the datatype
type(y_train)
```

```python
# Plotting the decision boundary on training data
plot_decision_regions(X_train, y_train, clf=regressor, legend=2)
```

```python
# make predictions (assign class labels)
y_pred = regressor.predict(X_test)
```

```python
# import confusion matrix library
from sklearn.metrics import confusion_matrix
# perform confusion matrix and getting accuracy score
con=sklearn.metrics.confusion_matrix(Y_test,y_pred)
print(con)
print(sklearn.metrics.accuracy_score(np.array(Y_test),y_pred))
```

```python
# Plotting the decision boundary from the actual test data to compare with predicted data
plot_decision_regions(X_test, y_test, clf=regressor, legend=2)
```

```python
# Plotting the new decision boundary with predicted data
# from the LogisticRegression model
plot_decision_regions(X_test, y_pred, clf=regressor, legend=2)
```

The Logistic Regression model has been implemented on Wine data set. The model provides an accuracy of around 86% in prediction using test data.
